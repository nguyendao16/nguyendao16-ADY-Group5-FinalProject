```python
import pandas as pd
from dotenv import load_dotenv
import os
load_dotenv()

URL = os.getenv("URL")

df = pd.read_csv("./Strokesdataset.csv")
df.groupby("stroke")["stroke"].agg(len)
```

```python
df.loc[
    (df["smoking_status"].isna()) &
    (df["work_type"] != "children") &
    (df["age"] < 18),
    "smoking_status"
] = "never smoked"

df.loc[
    df["work_type"] == "children",
    "smoking_status"
] = "never smoked"
```
